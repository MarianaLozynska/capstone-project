```python
import numpy as np
```

```python
base = "../initial_data"
inputs = {}
outputs = {}

for i in range(1, 9):
    folder = f"{base}/function_{i}"
    inputs[i] = np.load(f"{folder}/initial_inputs.npy")
    outputs[i] = np.load(f"{folder}/initial_outputs.npy")
```

```python
week1_inputs = {
    1: np.array([0.4, 0.4]),
    2: np.array([0.72, 0.90]),
    3: np.array([0.51, 0.63, 0.32]),
    4: np.array([0.60, 0.45, 0.41, 0.27]),
    5: np.array([0.24, 0.83, 0.90, 0.90]),
    6: np.array([0.75, 0.17, 0.71, 0.67, 0.08]),
    7: np.array([0.08, 0.47, 0.27, 0.24, 0.44, 0.76]),
    8: np.array([0.07, 0.08, 0.04, 0.06, 0.42, 0.82, 0.51, 0.88])
}

week1_outputs = {
    1: np.float64(0.09787028091182691),
    2: np.float64(0.5567314070917742),
    3: np.float64(-0.05929363571874033),
    4: np.float64(-4.416289141043837),
    5: np.float64(1231.606836011881),
    6: np.float64(-0.5919788840023106),
    7: np.float64(1.364636869415445),
    8: np.float64(9.58626)
}
```

```python
combined_inputs = {}
combined_outputs = {}

for i in range(1, 9):
    combined_inputs[i] = np.vstack([inputs[i], week1_inputs[i]])
    combined_outputs[i] = np.append(outputs[i], week1_outputs[i])
```
